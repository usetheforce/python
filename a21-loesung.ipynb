```python
import sqlite3 
datenbankverbindung = sqlite3.connect( "terra1.db") 
cursor = datenbankverbindung.cursor()

cursor.execute('SELECT Name,Land,Einwohner FROM ort WHERE Name LIKE "%dorf%" ORDER BY Einwohner ASC;') 

ergebnis = cursor.fetchall()
datenbankverbindung.close() 

print( "Es gibt",len(ergebnis),"Orte:" )
einwohnerGesamt = 0
for datum in ergebnis:
    print( datum[0],",",datum[1],",",datum[2],"Einwohner" )
    einwohnerGesamt = einwohnerGesamt + datum[2]
print( "Einwohner gesamt:",einwohnerGesamt )
```
